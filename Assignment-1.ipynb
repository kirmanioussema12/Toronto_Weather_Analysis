# <span style="color:Green">P1: Load the dataset.</span>

```python
import csv
file_path=r"C:\Users\MSI\Desktop\assignment 1\TORONTO_INTL_A_Climate_Daily_Data_2025 (1).csv"
```

```python
#loading the Dataset using the csv module which is a built-in module in Python for handling CSV files.
# We open the file in read mode with UTF-8 encoding, create a DictReader to read the CSV data into a list of dictionaries,
# and then print the first 5 rows of the dataset to verify that it has been loaded correctly.
data = []
with open(file_path, "r", encoding="utf-8-sig") as file:
    reader = csv.DictReader(file)
    for row in reader:
        data.append(row)
print(data[:5])


```

# <span style="color:Green">P2: What is the overall summary (mean, median, min, max, standard deviation) of temperature, precipitation, and wind speed for the year?</span>

```python
import math
#the function to_float is designed to convert a string input into a floating number while handling cases where the input might be empty or represnet is"nan""

def to_float(x):
    x = x.strip()
    if x == "" or x.lower() == "nan":
        return None
    return float(x)
#the function get_column_as_floats takes a list of dictionaries (data) and a column name, and it extracts the values from that column, 
#converts them to floats using the to_float function, and returns a list of valid float values.

def get_column_as_floats(data, column_name):
    values = []
    for row in data:
        value = to_float(row[column_name])
        
        if value is not None:
            values.append(value)
    return values

#calculating summary statistics for the specified columns (Mean Temp, Total Precip, Spd of Max Gust)
# using the defined functions for mean, median, and sample standard deviation.

def mean(values):
    return sum(values) / len(values)

def median(values):
    sorted_vals = sorted(values)
    n = len(sorted_vals)
    mid = n // 2

    if n % 2 == 1:
        return sorted_vals[mid]
    else:
        return (sorted_vals[mid - 1] + sorted_vals[mid]) / 2

def sample_std_dev(values):
    n = len(values)
    if n < 2:
        return 0.0

    m = mean(values)
    variance = sum((x - m) ** 2 for x in values) / (n - 1)   # sample variance
    return math.sqrt(variance)
#using the round function to round the computed statistics to two decimal places for better readability.

def summary_stats(values):
    return {
        "mean": round(mean(values), 2),
        "median": round(median(values), 2),
        "min": min(values),
        "max": max(values),
        "std_dev": round(sample_std_dev(values), 2)
    }

# Extract columns
temps = get_column_as_floats(data, "Mean Temp (°C)")
precip = get_column_as_floats(data, "Total Precip (mm)")
wind = get_column_as_floats(data, "Spd of Max Gust (km/h)")

# Compute stats
temp_stats = summary_stats(temps)
precip_stats = summary_stats(precip)
wind_stats = summary_stats(wind)

# Print results
print(" Summary of Temperature (Mean Temp °C) ")
print(temp_stats)

print("\n Summary of Precipitation (Total Precip mm) ")
print(precip_stats)

print("\n Summary of Wind Speed (Spd of Max Gust km/h) ")
print(wind_stats)
```

# <span style="color:Green">P3: What are the highest and lowest recorded temperatures of the year, and on which dates did they occur?</span>

```python

# P3: highest and lowest recorded temperatures + their dates
#i'm taking the highest temperature from the "Max Temp (°C)" column and the lowest temperature from the "Min Temp (°C)" column,
# and not using the mean temperature for that day 
# because the maximum of the temperature is not the mean of a day but it's the maximum of the day and the minimum is the minimum of the day, 
# so it makes more sense to use those columns to find the highest and lowest temperatures recorded in the year.


highest_temp = None
highest_date = None

lowest_temp = None
lowest_date = None

for row in data:
    date = row["Date/Time"]

    max_temp = to_float(row["Max Temp (°C)"])
    min_temp = to_float(row["Min Temp (°C)"])

    # Highest temperature of the year (from Max Temp column)
    if max_temp is not None:
        if highest_temp is None or max_temp > highest_temp:
            highest_temp = max_temp
            highest_date = date

    # Lowest temperature of the year (from Min Temp column)
    if min_temp is not None:
        if lowest_temp is None or min_temp < lowest_temp:
            lowest_temp = min_temp
            lowest_date = date

print("Highest recorded temperature:", highest_temp, "°C on", highest_date)
print("Lowest recorded temperature:", lowest_temp, "°C on", lowest_date)
```

# <span style="color:Green">P4: What is the total annual precipitation (rain + snow)?</span>

```python
#Total annual precipitation (rain + snow)




total_rain_mm = 0.0
total_snow_mm = 0.0
# Rain is in mm, Snow is in cm

for row in data:
    rain = row["Total Rain (mm)"]
    snow = row["Total Snow (cm)"]

    rain_val = to_float(rain) if rain is not None else None
    snow_val = to_float(snow) if snow is not None else None

    if rain_val is not None:
        total_rain_mm += rain_val

    if snow_val is not None:
        total_snow_mm += snow_val * 10   # Convert snow: 1 cm snow = 10 mm water equivalent

total_precip_mm = total_rain_mm + total_snow_mm

print("Total annual rain:", round(total_rain_mm, 2), "mm")
print("Total annual snow (converted):", round(total_snow_mm, 2), "mm")
print("Total annual precipitation (rain + snow):", round(total_precip_mm, 2), "mm")
```

# <span style="color:Green">P5: How many days had snowfall, and what was the total snowfall for the year?</span>

```python
#i'm counting the number of days with snowfall and the total snowfall for the year by iterating through each row of the dataset,
#checking the "Total Snow (cm)" column, and summing up the values while counting the days with snowfall.
snow_days = 0
total_snow_cm = 0.0

for row in data:
    snow = row["Total Snow (cm)"]
    snow_val = to_float(snow) if snow is not None else None

    if snow_val is not None and snow_val > 0:
        snow_days += 1
        total_snow_cm += snow_val

print("Number of days with snowfall:", snow_days)
print("Total snowfall for the year:", round(total_snow_cm, 2), "cm")
```

# <span style="color:Green">P6: What is the average temperature for each month?</span>

```python

monthly_sum = {}     # month -> sum of temps
monthly_count = {}   # month -> number of valid temp values
for row in data:
    month = int(row["Month"])
    temp = to_float(row["Mean Temp (°C)"])

    if temp is not None:
        if month not in monthly_sum:
            monthly_sum[month] = 0.0
            monthly_count[month] = 0

        monthly_sum[month] += temp
        monthly_count[month] += 1

print("Average temperature for each month (Mean Temp °C):")
for month in range(1, 13):
    if month in monthly_sum:
        avg = monthly_sum[month] / monthly_count[month]
        print("Month", month, ":", round(avg, 2), "°C")
    else:
        print("Month", month, ": No data")
```

# <span style="color:Green">P7: What are the temperature, precipitation, and wind speed during each season: Winter, Spring, Summer, and Fall? </span>

```python
# P7: Seasonal temperature, precipitation, and wind speed

import math

seasons = {1: 'Winter', 2: 'Winter', 3: 'Spring', 4: 'Spring', 5: 'Spring',
           6: 'Summer', 7: 'Summer', 8: 'Summer', 9: 'Fall', 10: 'Fall', 11: 'Fall'}



# Store sums and counts for averages
season_temp_sum = {"Winter": 0.0, "Spring": 0.0, "Summer": 0.0, "Fall": 0.0}
season_temp_daily = {"Winter": 0, "Spring": 0, "Summer": 0, "Fall": 0}

season_wind_sum = {"Winter": 0.0, "Spring": 0.0, "Summer": 0.0, "Fall": 0.0}
season_wind_daily = {"Winter": 0, "Spring": 0, "Summer": 0, "Fall": 0}

season_precip_sum = {"Winter": 0.0, "Spring": 0.0, "Summer": 0.0, "Fall": 0.0}
season_precip_daily = {"Winter": 0, "Spring": 0, "Summer": 0, "Fall": 0}

# Loop through all the records and aggregate data by season
#i'm ignoring December (month 12) since it belongs to the next year's winter season and we want to keep the analysis within a single calendar year.
for row in data:
    month = int(row["Month"])
    if month==12:
        continue
    season = seasons[month]

    temp = to_float(row["Mean Temp (°C)"])
    precip = to_float(row["Total Precip (mm)"])
    wind = to_float(row["Spd of Max Gust (km/h)"])

    # Temperature (average)
    if temp is not None:
        season_temp_sum[season] += temp
        season_temp_daily[season] += 1

    # Wind (average)
    if wind is not None:
        season_wind_sum[season] += wind
        season_wind_daily[season] += 1

    # Precipitation (total)
    if precip is not None:
        season_precip_sum[season] += precip
        season_precip_daily[season] += 1

# Print results
print("=== Seasonal Climate Summary ===\n")

for season in ["Winter", "Spring", "Summer", "Fall"]:
    avg_temp = season_temp_sum[season] / season_temp_daily[season] 
    avg_wind = season_wind_sum[season] / season_wind_daily[season]
    avg_precip = season_precip_sum[season] / season_precip_daily[season] 

    print(season)
    print("  Avg Temperature:", round(avg_temp, 2), "°C")
    print("  Avg Precipitation:", round(avg_precip, 2), "mm")
    print("  Avg Wind Speed:", round(avg_wind, 2), "km/h")
    
```

# <span style="color:Green">P8: What is the average monthly precipitation, and which month had the highest/lowest precipitation? </span>

```python
monthly_precip_sum = {}    # month -> total precipitation
monthly_precip_count = {}  # month -> number of valid days

for row in data:
    month = int(row["Month"])
    precip = to_float(row["Total Precip (mm)"])

    if precip is not None:
        if month not in monthly_precip_sum:
            monthly_precip_sum[month] = 0.0
            monthly_precip_count[month] = 0

        monthly_precip_sum[month] += precip
        monthly_precip_count[month] += 1

# Compute average monthly precipitation
monthly_avg_precip = {}

print("Average monthly precipitation (mm):")
for month in range(1, 13):
    if month in monthly_precip_sum:
        avg = monthly_precip_sum[month] / monthly_precip_count[month]
        monthly_avg_precip[month] = avg
        print("Month", month, ":", round(avg, 2), "mm")
    else:
        print("Month", month, ": No data")

# Find highest and lowest total precipitation month
highest_month = None
lowest_month = None

for month in monthly_precip_sum:
    if highest_month is None or monthly_precip_sum[month] > monthly_precip_sum[highest_month]:
        highest_month = month

    if lowest_month is None or monthly_precip_sum[month] < monthly_precip_sum[lowest_month]:
        lowest_month = month

print("\nMonth with highest precipitation:",
      highest_month, "with", round(monthly_precip_sum[highest_month], 2), "mm")

print("Month with lowest precipitation:",
      lowest_month, "with", round(monthly_precip_sum[lowest_month], 2), "mm")
```

# <span style="color:Green">P9: What is the correlation between temperature and precipitation? The Pearson Correlation formula:
$$r = \frac{\sum(x_i-\overline{x})(y_i-\overline{y})}{\sqrt{\sum{(x_i-\overline{x})}^2{\sum(y_i-\overline{y})}^2}}$$
</span>

```python


# Collect paired (x, y) values
x_vals = []   # temperatures
y_vals = []   # precipitation

for row in data:
    temp = to_float(row["Mean Temp (°C)"])
    precip = to_float(row["Total Precip (mm)"])

    if temp is not None and precip is not None:
        x_vals.append(temp)
        y_vals.append(precip)

n = len(x_vals)

# Compute means
x_mean = sum(x_vals) / n
y_mean = sum(y_vals) / n

# Compute numerator and denominator
numerator = 0.0
sum_x_diff_sq = 0.0
sum_y_diff_sq = 0.0

for i in range(n):
    x_diff = x_vals[i] - x_mean
    y_diff = y_vals[i] - y_mean

    numerator += x_diff * y_diff
    sum_x_diff_sq += x_diff ** 2
    sum_y_diff_sq += y_diff ** 2

denominator = math.sqrt(sum_x_diff_sq * sum_y_diff_sq)

# Pearson correlation coefficient
r = numerator / denominator if denominator != 0 else 0.0

print("Pearson correlation (Temperature vs Precipitation):", round(r, 4))
```

# <span style="color:Green">P10: How many days recorded extreme weather conditions, such as temperatures exceeding 30°C or dropping below -20°C? </span>

```python
# P10: Extreme weather days (> 30°C or < -20°C)


extreme_count = 0
extreme_days = []  # store dates for reference

for row in data:
    date = row["Date/Time"]
    max_temp = to_float(row["Max Temp (°C)"])
    min_temp = to_float(row["Min Temp (°C)"])

    is_hot = (max_temp is not None and max_temp > 30.0)
    is_cold = (min_temp is not None and min_temp < -20.0)

    if is_hot or is_cold:
        extreme_count += 1
        extreme_days.append(date)

print("Number of extreme weather days:", extreme_count)


```
